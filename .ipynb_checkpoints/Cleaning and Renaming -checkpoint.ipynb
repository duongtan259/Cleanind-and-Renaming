Import necessary libraries

```python
import pandas as pd
import numpy as np
```

Read the Excel file into Dataframe

```python
df = pd.read_excel("all_table_definitions.xlsx")
df
```

Rename the column of the Dataframe

```python
df.rename(columns={'Field Name     | Description                                                                                 | Domain          | Datatype     Physic.| Mand. | Acti  | Refer.   Reference':'General'}, inplace=True)
```

```python
df
```

Filter the data to get only the necessary rows

```python
list_numbers = str([1, 2, 3, 4, 5, 6, 7, 8, 9])
df1 = df[df['General'].astype(str).str.startswith(tuple(list_numbers))]
df1
```

Split the data in to the readable columns

```python
df2 = df1.copy()
df2[['Field Name', 'Description', 'Domain', 'Datatype Physic.', 'Mand.',  'Acti',  'Refer. Reference']] = df2['General'].str.split('|', expand=True)
df2.drop(columns='General', inplace=True)
df2
```

Cleaning up the number in the Field Name column

```python
df2['Field Name'] = df2['Field Name'].str.replace(list_numbers, "")
df2.head()
```

Cleaning the bank rows in 2 columns

```python
Clean_columns = ['Field Name', 'Description']

for column in Clean_columns:
    df2[column] = df2[column].str.strip()
    df2[column].replace('', np.nan, inplace=True)
    df2 = df2[df2[column].notna()]
    
df2
```

Remove the duplicate values in 2 columns 

```python
df2 = df2.drop_duplicates(
  subset = ['Field Name', 'Description'],
  keep = 'last').reset_index(drop = True)
df2
```

Save the data to Excel file

```python
df2.to_excel('Dataset.xlsx')
```
